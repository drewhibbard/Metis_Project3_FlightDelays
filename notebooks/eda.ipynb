# Flights - Exploratory Data Analysis

This notebook is exploratory data analysis only.  I actually ended up replicating all of this and more in Tableau.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
flights = pd.read_csv('flights_final.csv',index_col=0)
```

```python
flights.columns
```

```python
delays_by_airline = flights.groupby('airline_name').sum()['delayed']
flights_by_airline = flights.groupby('airline_name').count()['tail_num']
percent_delay_airline = delays_by_airline/flights_by_airline
```

```python
plt.barh(percent_delay_airline.index,percent_delay_airline.sort_values())
plt.xlim(0,1)
plt.title('Percent of Flights Delayed by Airline');
```

```python
delays_by_age = flights.groupby('aircraft_age').sum()['delayed']
total_by_age = flights.groupby('aircraft_age').count()['tail_num']
percent_delay_age = delays_by_age/total_by_age
```

```python
plt.bar(percent_delay_age.index,percent_delay_age)
plt.ylim(0,1)
plt.title('Percent of Flights Delayed by Age of Aircraft')
plt.xlabel('Aircraft Age (Years)');
```

```python
delays_by_airport = flights.groupby('airport_name_orig').sum()['delayed']
total_flights_airport = flights.groupby('airport_name_orig').count()['tail_num']
percent_delay_airport = delays_by_airport/total_flights_airport
```

```python
plt.barh(percent_delay_airport.index[-5:],percent_delay_airport.sort_values().tail())
plt.title('Percent of Flights Delayed by Departing Airport (5 Worst)',fontsize=20)
plt.xlim(0,1);
```

```python
plt.barh(percent_delay_airport.index[:5],percent_delay_airport.sort_values().head())
plt.title('Percent of Flights Delayed by Departing Airport (5 Best)',fontsize=20)
plt.xlim(0,1);
```

```python
delays_by_month = flights.groupby('month').sum()['delayed']
total_flights_month = flights.groupby('month').count()['tail_num']
percent_delay_month = delays_by_month/total_flights_month
```

```python
plt.bar(percent_delay_month.index,percent_delay_month)
plt.title('Percent of Flights Delayed by Month',fontsize=20)
plt.ylim(0,1);
```

```python
delays_by_hour = flights.groupby('hour').sum()['delayed']
total_flights_hour = flights.groupby('hour').count()['tail_num']
percent_delay_hour = delays_by_hour/total_flights_hour
```

```python
plt.bar(percent_delay_hour.index,percent_delay_hour)
plt.title('Percent of Flights Delayed by Hour of Departure',fontsize=18)
plt.ylim(0,1);
```

```python
delays_by_weekday = flights.groupby('day_of_week').sum()['delayed']
total_flights_weekday = flights.groupby('day_of_week').count()['tail_num']
percent_delay_weekday = delays_by_weekday/total_flights_weekday
```

```python
plt.bar(percent_delay_weekday.index,percent_delay_weekday)
plt.title('Percent of Flights Delayed by Day of Week',fontsize=18)
plt.xticks(ticks=range(1,8),labels=['Mon','Tu','Wed','Th','Fri','Sat','Sun'])
plt.ylim(0,1);
```

```python
delays_by_engine = flights.groupby('engine_manufacturer').sum()['delayed']
total_flights_engine = flights.groupby('engine_manufacturer').count()['tail_num']
percent_delay_engine = delays_by_engine/total_flights_engine
```

```python
plt.barh(percent_delay_engine.index,percent_delay_engine.sort_values())
plt.title('Percent of Flights Delayed by Engine Manufacturer',fontsize=18)
plt.xlim(0,1)
plt.yticks(range(0,6),labels=['Rolls Royce','Pratt & Whitney','International Aero Engines',
                             'General Electric','CFM International','BMW']);
```

```python
flights['snow_bin'] = flights.snow_orig>0
```

```python
delays_by_snow = flights.groupby('snow_bin').sum()['delayed']
total_flights_snow = flights.groupby('snow_bin').count()['tail_num']
percent_delay_snow = delays_by_snow/total_flights_snow
```

```python
plt.bar(percent_delay_snow.index,percent_delay_snow)
plt.title('Percentage of Flights Delayed - Snow Day Of',fontsize=18)
plt.ylim(0,1)
plt.xticks(range(0,2),labels=['No','Yes']);
```

```python
flights['precip_bin'] = flights.precip_orig>0

delays_by_precip = flights.groupby('precip_bin').sum()['delayed']
total_flights_precip = flights.groupby('precip_bin').count()['tail_num']
percent_delay_precip = delays_by_precip/total_flights_precip
```

```python
plt.bar(percent_delay_precip.index,percent_delay_precip)
plt.title('Percent of Flights Delayed - Precipitation Day Of',fontsize=18)
plt.ylim(0,1)
plt.xticks(range(0,2),labels=['No','Yes']);
```
